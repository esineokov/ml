```python
X = df.drop(columns=['Price'])
X = pd.get_dummies(X)
```

```python
y = df['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.3, random_state=42)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```
