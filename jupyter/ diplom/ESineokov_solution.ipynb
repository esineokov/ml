```python
def evaluate_preds(true_values, pred_values):
    print("R2:\t" + str(round(r2(true_values, pred_values), 3)) + "\n" +
          "MAE:\t" + str(round(mae(true_values, pred_values), 3)) + "\n" +
          "MSE:\t" + str(round(mse(true_values, pred_values), 3)))
    
    plt.figure(figsize=(10,10))
    
    sns.scatterplot(x=pred_values, y=true_values)
    
    plt.xlabel('Predicted values')
    plt.ylabel('True values')
    plt.title('True vs Predicted values')
    plt.show()
```

##### LifeSquare

```python
# Нулевые значения жилой площади заменим медианой
df.loc[df['LifeSquare'].isnull(), 'LifeSquare'] = df['LifeSquare'].median()
```

#### Healthcare_1

```python
# Нулевые значения Healthcare_1 заменим медианой
df.loc[df['Healthcare_1'].isnull(), 'Healthcare_1'] = df['Healthcare_1'].median()
```

### Обработка выбросов

```python
df.describe()
```

#### Rooms

```python
# Поскольку в задаче сказано, что к нежвижимости относятся только квартиры, 
# то считаем, что 0 комнат - это выброс, требующий исправления

# Приравняем такие значения к наиболее часто встречающемуся значению
df.loc[df['Rooms'] == 0.0, 'Rooms'] = df.loc[df['Rooms'] != 0.0, 'Rooms'].mode()[0]
```

#### KitchenSquare

```python
# Из вывода метода df.describe было видно, что KitchenSquare содержит нулевые значения
# Кухни с 0 площадью заменим медианой
df.loc[df['KitchenSquare'] == 0.0, 'KitchenSquare'] = df.loc[df['KitchenSquare'] != 0.0, 'KitchenSquare'].median()
```

```python
# Из вывода метода df.describe было видно, что у KitchenSquare слишком большое значение max
# Оценим визуально как много таких значений
df.sort_values(by=['KitchenSquare'])
```

```python
# Два последних значения KitchenSquare явная ошибка с полем HouseYear
# Заменим эти ошибки разницей общей площади квартиры и жилой площади
df.loc[df['KitchenSquare'] > 1000, 'KitchenSquare'] = df.loc[df['KitchenSquare'] > 1000, 'Square'] - df.loc[df['KitchenSquare'] > 1000, 'LifeSquare']
```

#### LifeSquare

```python
# Из вывода метода df.describe было видно, что у LifeSquare слишком большое значение max
# Оценим визуально как много таких значений
df.sort_values(by=['LifeSquare'])
```

```python
# Значение LifeSquare у последней записи слишком большое.
# Заменим его разницей между общей площадью и площадью кухни 
df.loc[df['LifeSquare'] > 1000, 'LifeSquare'] = df.loc[df['LifeSquare'] > 1000, 'Square'] - df.loc[df['LifeSquare'] > 1000, 'LifeSquare']
```

#### Square

```python
# Проверим, что общая площадь квартиры не меньше, чем сумма жилой площади и площади кухни
df.loc[df['Square'] < df['LifeSquare'] + df['KitchenSquare']]
```

```python
df.sort_values(by=['Square']).tail(10)
```

#### HouseFloor

```python
# Последние 3 записи слишком неправдоподобны по своим данным и выбиваются от остальных. 
# Данные квартиры с максимальной площадьью имеют всего по 1-2 комнаты и цены ниже остальных.
# Удалим эти записи.
df.drop(index=[4690, 4262, 6977], inplace=True)
```

```python
# Из вывода метода df.describe было видно, что у HouseFloor есть нулевые значения
# Правняем нулевые значения к этажу квартиры
df.loc[df['HouseFloor'] == 0.0, 'HouseFloor'] = df.loc[df['HouseFloor'] == 0.0, 'Floor']
```

#### Floor

```python
# Убедимся, что этаж квартиры не больше этажей всего дома
df.loc[df['HouseFloor'] < df['Floor']]
```

```python
# 1) Правняем такие значения к этажу квартиры
df.loc[df['HouseFloor'] < df['Floor'], 'HouseFloor'] = df.loc[df['HouseFloor'] < df['Floor'], 'Floor']
```

#### HouseYear

```python
# Всем дома, год построки которых больще текущего, установм год, который чаще всего повторяется в датасете 
df.loc[df['HouseYear'] > 2020] = df['HouseYear'].mode()[0]
```

### Все предыдущие операции по нормальзации данных применим и к тестовому датасету

```python
df_test = pd.read_csv(TEST_DATASET_PATH)
df_test.set_index('Id', inplace=True)
df_test.loc[df_test['LifeSquare'].isnull(), 'LifeSquare'] = df_test['LifeSquare'].median()
df_test.loc[df_test['Healthcare_1'].isnull(), 'Healthcare_1'] = df_test['Healthcare_1'].median()
df_test.loc[df_test['Rooms'] == 0.0, 'Rooms'] = df_test.loc[df_test['Rooms'] != 0.0, 'Rooms'].mode()[0]
df_test.loc[df_test['KitchenSquare'] == 0.0, 'KitchenSquare'] = df_test.loc[df_test['KitchenSquare'] != 0.0, 'KitchenSquare'].median()
df_test.loc[df_test['KitchenSquare'] > 1000, 'KitchenSquare'] = df_test.loc[df_test['KitchenSquare'] > 1000, 'Square'] - df_test.loc[df_test['KitchenSquare'] > 1000, 'LifeSquare']
df_test.loc[df_test['LifeSquare'] > 1000, 'LifeSquare'] = df_test.loc[df_test['LifeSquare'] > 1000, 'Square'] - df_test.loc[df_test['LifeSquare'] > 1000, 'LifeSquare']
df_test.loc[df_test['Square'] < df_test['LifeSquare'] + df_test['KitchenSquare']]
df_test.loc[df_test['HouseFloor'] == 0.0, 'HouseFloor'] = df_test.loc[df_test['HouseFloor'] == 0.0, 'Floor']
df_test.loc[df_test['HouseFloor'] < df_test['Floor'], 'HouseFloor'] = df_test.loc[df_test['HouseFloor'] < df_test['Floor'], 'Floor']
df_test.loc[df_test['HouseYear'] > 2020] = df_test['HouseYear'].mode()[0]
df_test = pd.get_dummies(df_test, columns=['Ecology_2', 'Ecology_3', 'Shops_2'])
df_test.info()
```

### Разбиение на train и test

```python
X = df[feature_names]
y = df[target_name]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, shuffle=True, random_state=21)
```

### Построение базовых моделей и выбор лучшей

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
```

```python
y_train_preds = lr_model.predict(X_train)
evaluate_preds(y_train, y_train_preds)
```

```python
y_test_preds = lr_model.predict(X_test)
evaluate_preds(y_test, y_test_preds)
```

**Random Forest**

```python
rf_model = RandomForestRegressor()
rf_model.fit(X_train, y_train)
```

```python
y_train_preds = rf_model.predict(X_train)
evaluate_preds(y_train, y_train_preds)
```

```python
y_test_preds = rf_model.predict(X_test)
evaluate_preds(y_test, y_test_preds)
```

**Gradient Boosting**

```python
gb_model = GradientBoostingRegressor()
gb_model.fit(X_train, y_train)
```

```python
y_train_preds = gb_model.predict(X_train)
evaluate_preds(y_train, y_train_preds)
```

```python
y_test_preds = gb_model.predict(X_test)
evaluate_preds(y_test, y_test_preds)
y_test_preds
```
